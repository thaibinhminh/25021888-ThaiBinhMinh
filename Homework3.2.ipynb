```python
#Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím 
h=int(input())
y=0
x=h-1
for i in range(1,h+1):
    if i==(h):
        print('*'*((h*2)-1))
    else:
        x1=' '*x
        y1=' '*y
        if i==1:
            print(x1+'*')
            y+=1
        else:
            print(x1+'*'+y1+'*')
            y+=2
        x-=1
        
```

```python
#Nhập vào số nguyên dương a, in ra tất cả các ước dương của a 
import math
a=int(input())
for i in range(1, int(math.sqrt(a))+1):
    if a%i==0:
        print(i, end=' ')
        x=a//i
        if x!=i:
            print(x, end=' ')
    
```

```python
#Nhập vào số nguyên dương a, đếm số ước dương của a
import math
a=int(input())
cnt=0
for i in range(1, int(math.sqrt(a))+1):
    if a%i==0:
        cnt+=1
        x=a//i
        if x!=i:
            cnt+=1
print(cnt)
```

```python
#Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b 
import math
a, b=map(int,input().split())
for i in range(1, int(math.sqrt(a))+1):
    if a%i==0:
        if b%i==0:
            print(i, end=' ')
        x=a//i
        if x!=i and b%x==0:
            print(x, end=' ')
        
        
```

```python
#Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
#Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 
a=int(input())
s=0
for i in range(1,int(a**0.5)+1):
    if a%i==0:
        s+=i
        x=a//i
        if x!=i and x!=a:
            s+=x
if s==a:
    print('là số hoàn hảo')
else:
    print('không phải là số hoàn hảo')
```

```python
#Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 
A=int(input())
s=0
n=1
while s<=A:
    s+=1/n
    n+=1
print(n-1)
```

```python
#Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. 
#Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng
n=int(input())
max1=n
min1=n
while n!=-1:
    n=int(input())
    if max1<n:
        max1=n
    if min1>n and n!=(-1):
        min1=n
print(f'Số lớn nhất {max1}, số bé nhất {min1}')
    
```

```python
#Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 
n=int(input())
chu_so_chan=0
chu_so_le=0
while n>0:
    x=n%10
    n=n//10
    if x%2==0:
        chu_so_chan+=1
    else:
        chu_so_le+=1
print(f'Có {chu_so_chan} chữ số chẵn và {chu_so_le} chữ số lẻ')
    
```

```python
#Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay không 
n=int(input())
while n>1 and n%3==0:
    n=n//3
if n==1:
    print('có dạng 3^k')
else:
    print('không có dạng 3^k')
    
```

```python
#Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
#kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
#dãy số fibonacci lớn nhất nhưng không vượt quá A
A=int(input())
a=1
b=1
c=2
while b<=A:
    a, b= b, c
    c=a+b
print(a)
```

```python
#Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố. 
n=int(input())
def so_nguyen_to(a):
    if a==1:
        return False
    else:
        for i in range(2,int(a**0.5)+1):
            if a%i==0:
                return False
    return True
cnt=0
while n>0:
    x=n%10
    if so_nguyen_to(x):
        cnt+=1
    n=n//10
print(cnt)
```
