```python
#Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu mỗi phần tử
#x trong mảng xuất hiện đúng x lần. Kiểm tra xem mảng có phải là mảng đẹp hay không.
#Nếu không hãy xóa đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp.
from collections import Counter
lst=list(map(int,input().split()))
cnt_dic=dict(Counter(lst))
check=True
cnt=0
for i in cnt_dic.keys():
    if cnt_dic[i]>i:
        cnt+=(cnt_dic[i]-i)
        check=False
    elif cnt_dic[i]<i:
        cnt+=cnt_dic[i]
        check=False
if check:
    print('YES')
else:
    print('NO', cnt)
```

```python
#Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ phức tạp thời
#gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore. Nếu không có
#phần tử nào thỏa mãn thì không in gì
lst=list(map(int,input().split()))
#Phân tích: Nếu có tồn tại giá trị xuất hiện nhiều hơn n/2 lần thì số lần xuất hiện của
#nó lớn hơn hẳn tổng số lần xuất hiện của các số còn lại ta xét xét để tìm số có thể 
#là số đó. Ta thấy khi số lần xuất hiện của nó lớn hơn hẳn tổng tất cả các lần xuất 
#hiện của số khác thì ta có sẽ tạo một biến bằng để khi gặp nó thì +1 và khi gặp số 
#khác nó thì -1, nếu nó bị trừ về 0 thì số đó không phải số thỏa mãn nên ta sẽ dổi số 
#khác. Vì độ phức tạc thời gian cho phép là O(n) nên ta sẽ code như sau:
candidate=None
cnt=0
for i in lst:
    if cnt==0:
        candidate=i
        cnt=1
    elif i==candidate:
        cnt+=1
    else:
        cnt-=1
#Đến đây thì candidate chưa chắc là đáp án chính xác nên ta phải kiểm tra lại, nhưng 
#trước khi kiểm tra lại thì ta nên có một bước phòng thủ là kiểm tra xem nếu input
#là rỗng thì candidate = none để chắc chắn thì ta nên có một bước kiểm tra
if candidate is not None:
    cnt2=0
    for x in lst:
        if x == candidate:
            cnt2+=1
    if cnt2>len(lst)//2:
        print(candidate)
```

```python
#Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ hay không
def giai(n):
    my_dic={'}':'{', ']':'[', ')':'('}
    kiem_tra=[]
    for i in n:
        if i in '{[(':
            kiem_tra.append(i)
        else:
            if not kiem_tra:
                return False
            if kiem_tra[-1]!=my_dic[i]:
                return False
            else:
                kiem_tra.pop(-1)
    if len(kiem_tra)==0:
        return True
    return False
n=input()
if giai(n):
    print('YES')
else:
    print('NO')
```

```python
#Bài toán Tìm dãy các hợp số dài nhất liên tiếp chứa số nguyên < N (N
#là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ dài thì lấy dãy nhỏ
#nhất (gặp đầu tiên từ trái qua)
def snt(a):
    if a == 1:
        return False
    for i in range(2,int(a**0.5)+1):
        if a%i==0:
            return False
    return True
N=int(input())
lst=[(0,0)]
i=1
while i<N:
    cnt=0
    x=[]
    while not snt(i) and 1<i<N:
        cnt+=1
        x.append(i)
        i+=1
    if cnt>lst[0][0]:
        lst[0]=(cnt,x)
    i+=1
print(*lst[0][1])
```

```python
#Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ bàn
#phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau
a=list(map(int,input().split()))
best_cnt=0
lst=[]
cnt=1
x=[]
for i in range(len(a)-1):
    if a[i+1]>a[i] and cnt==1:
        cnt+=1
        x.append(a[i])
        x.append(a[i+1])
    elif a[i+1]>a[i]:
        cnt+=1
        x.append(a[i+1])
    else:
        if cnt>best_cnt:
            best_cnt=cnt
            lst=list(x)
        cnt=1
        x=[]
if cnt>best_cnt:
    lst=list(x)
print(*lst)
```

```python
#Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập vào từ bàn phím
a=list(map(int,input().split()))
lst=[1]*len(a)
ptu_truoc=[-1]*len(a)
for i in range(len(a)):
    for j in range(0,i):
        if a[j]<=a[i] and lst[j]+1>lst[i]:
            lst[i]=lst[j]+1
            ptu_truoc[i]=j
max_len=max(lst)
idx=lst.index(max_len)

lst2=[]
while idx!=-1:
    lst2.append(a[idx])
    idx=ptu_truoc[idx]

lst2.reverse()
print(*lst2)
```

```python
#Đổi chỗ đúng 2 chữ số để được số nhỏ nhất có thể
def min_idx(arr):
    min_id=0
    for i in range(len(arr)-1,-1,-1):
        if arr[i]<arr[min_id]:
            min_id=i
    cnt=1
    while min_id==cnt-1:
        min_id=cnt
        if cnt<len(arr):
            for i in range(cnt,len(arr)):
                if arr[i]<arr[min_id]:
                    min_id=i
            cnt+=1
        else: return False   
    return min_id, cnt
a=input()
A=[]
for i in a:
    A.append(i)
kq_ham=min_idx(A)
if kq_ham is False:
    A[-1], A[-2]= A[-2], A[-1]
else:
    min_id, cnt=kq_ham
    A[min_id], A[cnt-1]=A[cnt-1], A[min_id]
kq=''
for i in A:
    kq+=i
print(int(kq))
```

```python
#Đổi chỗ đúng 2 chữ số để được số nhỏ nhất có thể
def giai(arr):
    lst=list(arr)
    min_idx=len(arr)-1

    swap_i=-1
    swap_j=-1
    for i in range(len(arr)-2,-1,-1):
        if lst[i]<lst[min_idx]:
            min_idx=i
        elif lst[i]>lst[min_idx]:
            swap_i=i
            swap_j=min_idx
    if swap_i==-1:
        lst[-1], lst[-2]=lst[-2], lst[-1]
    else:
        lst[swap_j], lst[swap_i]=lst[swap_i], lst[swap_j]
    kq=''
    for i in lst:
        kq+=i
    return int(kq)
arr=input()
print(giai(arr))
```
