```python
from pathlib import Path
WORKDIR=Path('.')

name_file=input('nhap ten file cua ban: ')
path=WORKDIR/name_file
def reverse(path: str)->int:
    with open(path, 'r') as f:
        s= f.read().strip()
    reversed_s=s[::-1]
    return int(reversed_s)
print(reverse(path))
```

```python
from pathlib import Path
WORKDIR=Path('.')

name_file=input('nhap ten file cua ban: ')
path=WORKDIR/name_file
def canWinNim(path: str)->bool:
    with open(path, 'r') as f:
        s=int(f.read().strip())
        if s%4==0:
            return False
        else:
            return True
if canWinNim(path):
    print('Tôi thắng')
else:
    print('Bạn thắng')
```

```python
from pathlib import Path
WORKDIR=Path('.')

file_name1=input('Nhập tên file thứ nhất: ')
path1=WORKDIR/file_name1
file_name2=input('Nhập tên file thứ hai: ')
path2=WORKDIR/file_name2
def areAnagram(path1: str, path2: str):
    dic1={}
    with open(path1, 'r') as f:
        s1=f.read().lower()
        for i in s1:
            if i not in dic1 and i!=' ':
                dic1[i]=1
            elif i!=' ':
                dic1[i]+=1
    dic2={}
    with open(path2, 'r') as f:
        s2=f.read().lower()
        for i in s2:
            if i not in dic2 and i!=' ':
                dic2[i]=1
            elif i!=' ':
                dic2[i]+=1
    return dic1==dic2
print(areAnagram(path1, path2))
```

```python
from pathlib import Path
WORKDIR=Path('.')
def Fibonacci(n):
    if n==0:
        return 'a'
    if n==1:
        return 'b'
    return Fibonacci(n-1)+Fibonacci(n-2)
path=WORKDIR/'w11A8.txt'
n=int(input())
with open(path, 'w') as f:
    for i in range(0, n+1):
        f.write(Fibonacci(i)+'\n')
        print(Fibonacci(i))
```

```python
from pathlib import Path
path=Path('.')/'w11A9.csv'
def grade10(path: str):
    with open(path, 'r', encoding='utf-8') as f:
        s=f.readlines()
        for i in s:
            i=list(i.strip().split(','))
            if i[5]=='10':
                print(i[1]+' '+i[2])
grade10(path)
```
