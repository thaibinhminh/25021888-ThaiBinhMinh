```python
#Viết hàm tìm kiếm nhị phân trên một danh sách đã được sắp xếp tăng dần.
#Trả về chỉ số của phần tử hoặc -1 nếu không tìm thấy. 
#mẫu hàm: def binary_search(arr, left, right, target).
def binary_search(arr, left, right, target):
    while left<=right:
        mid=(left+right)//2
        if arr[mid]==target:
            return mid
        elif arr[mid]>target:
            right=mid-1
        else:
            left=mid+1
    return -1
arr=list(map(int,input().strip().split()))
target=int(input())
left=0
right=len(arr)-1
print(binary_search(arr, left, right, target))
```

```python
#Cho danh sách arr và một số nguyên x, đếm xem x xuất hiện bao nhiêu lần
#trong danh sách. def count_occurrences(arr, x) 
def count_occurrences(arr, x):
    cnt=0
    for i in arr:
        if i == x:
            cnt+=1
    return cnt
arr=list(map(int,input().strip().split()))
x=int(input())
print(count_occurrences(arr, x))
```

```python
#Viết hàm thực hiện thuật toán bubble sort, SelectionSort để sắp xếp danh sách 
#theo thứ tự tăng dần. Sắp xếp nổi bọt (Bubble Sort) Sắp xếp chọn (Selection Sort)
def bbsort(arr):
    n=len(arr)
    for i in range (n-1):
        swapp=False
        for j in range(0, n-1-i):
            if arr[j]>arr[j+1]:
                arr[j], arr[j+1]=arr[j+1], arr[j]
                swapp=True
        if not swapp:
            break
def slsort(arr):
    n=len(arr)
    for i in range(n-1):
        min_idx=i
        for j in range(i+1,n):
            if arr[j]<arr[min_idx]:
                min_idx=j
        arr[i], arr[min_idx]= arr[min_idx], arr[i]
arr=list(map(int,input().split()))
bbsort(arr)
print(*arr)
slsort(arr)
print(*arr)
```

```python
#Viết hàm tìm phần tử có số lần xuất hiện nhiều nhất trong danh sách. Nếu có
#nhiều phần tử cùng số lần, trả về phần tử xuất hiện trước.
def xuat_hien_nhieu(lst):
    dic={}
    for i in lst:
        if i not in dic:
            dic[i]=1
        else:
            dic[i]+=1
    key=lst[0]
    vl=1
    for i in dic:
        if dic[i]>vl:
            key=i
            vl=dic[i]
    return key, vl
lst=list(map(int,input().split()))
a, b=xuat_hien_nhieu(lst)
print(f'{a} xuat hien nhieu nhat, som nhat, {b} lan')
```

```python
def tong_x(lst,x):
    dic={}
    for i in lst:
        if i not in dic:
            dic[i]=1
        else:
            dic[i]+=1
    my_set=set()
    cnt=0
    for i in dic:
        j=x-i
        if (j in dic) and (i,j) not in my_set and i!=j:
            cnt+=(dic[i]*dic[j])
            my_set.add((j,i))
    return cnt
lst=list(map(int,input().split()))
x=int(input())
print(tong_x(lst,x))
```

```python
#dãy con tăng nghiêm ngặt dài nhất(O(n^2))
def dctdn(lst):
    lst1=[1]*len(lst)
    for i in range(1,len(lst)):
        for j in range(i):
            if lst[j]<lst[i]:
                lst1[i]=max(lst1[i],1+lst1[j])
    return max(lst1)
lst=list(map(int,input().split()))
print(dctdn(lst))
```

```python
#dãy con tăng nghiêm ngặt dài nhất(O(nlogn))
def gtbhlh(lst,x, l, r):
    while l<=r:
        mid=(l+r)//2
        if lst[mid]>=x:
            r=mid-1
        else:
            l=mid+1
    return l
def dctdn(lst):
    if not lst:
        return 0
    n=len(lst)
    lst1=[0]*n
    lst1[0]=lst[0]
    k=0
    for i in range(1, n):
        q=gtbhlh(lst1, lst[i], 0, k)
        if q>k:
            k+=1
            lst1[k]=lst[i]
        else:
            lst1[q]=lst[i]
    return k+1
lst=list(map(int,input().split()))
print(dctdn(lst))
```

```python
#Bạn được cung cấp: một danh sách các khoảng thời gian interval = [[l1, r1], [l2,
#r2], ...] và một danh sách các truy vấn queries = [q1, q2, ...]
#Với mỗi truy vấn q, hãy tìm độ dài nhỏ nhất của khoảng thời gian [li, ri] sao
#cho li ≤ q ≤ ri. Nếu không tồn tại khoảng thời gian nào như vậy, hãy
#trả về -1 cho truy vấn đó. solve in < O(m.n), hint: binary search in the queries list
import heapq
def giai(interval, queries):
    interval.sort()
    n=len(interval)
    m=len(queries)
    Q=[(queries[i],i) for i in range(m)]
    Q.sort()
    heap=[]
    ans=[-1]*m
    i=0
    for q, j in Q:
        while  i<n and interval[i][0]<=q:
            l, r= interval[i]
            heapq.heappush(heap, (r-l+1, r))
            i+=1
        while heap and heap[0][1] <q:
            heapq.heappop(heap)
        if heap:
            ans[j]=heap[0][0]        
    return ans
interval=eval(input())
queries=eval(input())
print(giai(interval, queries))
```

```python
from bisect import bisect_right
def giai(interval, queries):
    interval.sort()
    m=len(queries)
    ans=[]
    L=[interval[i][0] for i in range(m)]
    for i in queries:
        mn=10*9
        x=bisect_right(L, i) 
        if x==0:
            continue 
        for j in range(x):
            l, r= interval[j]
            if r>=i:
                mn=min(mn, r-l+1)
        if mn==10*9:
            ans.append(-1)
        else:
            ans.append(mn)
    return ans
interval=eval(input())
queries=eval(input())
print(giai(interval, queries))
```

```python
#Cho một danh sách các số nums, với mỗi phần tử nums[i], hãy đếm xem có
#bao nhiêu phần tử bên phải nhỏ hơn nums[i]. Trả về danh sách các số đếm. 
def giai(lst):
    n=len(lst)
    if n<=1:
        return lst
    mid=n//2
    l_L=lst[:mid]
    r_L=lst[mid:]
    l_L=giai(l_L)
    r_L=giai(r_L)
    return ho_tro(l_L, r_L)
def ho_tro(l_L, r_L):
    arr=[]
    i, j, m, n= 0, 0, len(l_L), len(r_L)
    while i<m and j<n:
        if l_L[i][0]>r_L[j][0]:
            arr.append(r_L[j])
            j+=1
        else:
            l_L[i][2]+=j
            arr.append(l_L[i])
            i+=1
    while i<m:
        l_L[i][2]+=j
        arr.append(l_L[i])
        i+=1
    arr.extend(r_L[j:])
    return arr
lst=eval(input())
x=len(lst)
L=[[lst[i], i, 0] for i in range(x)]
kq=[0]*x
L=giai(L)
for i in range(x):
    kq[L[i][1]]=L[i][2]
print(kq)
```

```python
from bisect import bisect_right, insort
lst=eval(input())
k=int(input())
n=len(lst)
s=[0]*n
s[0]=lst[0]%k
s_sx=[s[0]]
kq=s[0]
for i in range(1,n):
    s[i]=(s[i-1]+lst[i])%k
    kq=max(kq,s[i])
    x=bisect_right(s_sx,s[i])
    if x<len(s_sx):
        kq=max((s[i]-s_sx[x]+k)%k,kq)
    insort(s_sx, s[i])
print(kq)
```
