```python
#Nhập vào hai chuỗi a, b gồm các chữ cái thường. In ra "True" nếu chuỗi a dài hơn chuỗi b. Ngược lại in ra "False".
#Cách 1:
a=input('Nhập vào chuỗi a gồm các chữ cái thường: ')
b=input('Nhập vào chuỗi b gồm các chữ cái thường: ')
if a.isalpha and a.islower and b.isalpha and b.islower:
    if len(a)>len(b):
        print('True')
    else:
        print('False')
else:
    print('Vui lòng chỉ nhập kí tự thường')
print('------------------------------')
#Cách 2:
import re
if re.fullmatch(r'[a-z]+', a) and re.fullmatch(r'[a-z]+', b):
    if len(a)>len(b):
        print('True')
    else:
        print('False')
else:
    print('Vui lòng chỉ nhập kí tự thường')
print('------------------------------')
#Cách 3:
if all('a'<c<'z' for c in a) and all('a'<c<'z' for c in b):
    if len(a)>len(b):
        print('True')
    else:
        print('False')
else:
    print('Vui lòng chỉ nhập kí tự thường')
```

```python
#Kiểm tra 3 cạnh của tam giác:
a, b, c=map(float,input('Nhập độ dài ba cạnh tam giác cách nhau bằng dấu ",": ').split(','))
if a+b>c and b+c>a and a+c>b:
    if a==b==c:
        print('Đây là 3 cạnh của tam giác cân.')
    elif a**2+b**2==c**2 or c**2+b**2==a**2 or c**2+a**2==b**2:
        print('Đây là 3 cạnh của tam giác vuông.')
    elif a==b or a==c or b==c:
        print('Đây là 3 cạnh của tam giác cân')
    else:
        print('Đây là 3 cạnh của tam giác thường')
else:
    print('Đây không phải 3 cạnh của một tam giác')
```

```python
#Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 cấp số nhân với công bội nguyên theo đúng thứ tự a, b, c, d hay không? 
a, b, c, d=map(int,input('Nhâp 4 só theo đúng dạng a, b, c, d: ').split(','))
if b%a==0:
    q=b/a
    if b*q==c and c*q==d:
        print(f'{a},{b},{c},{d} là 1 cấp số nhân.')
else:
    print('Đây không phải là một cấp số nhân')
```
