```python
#Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ
#nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả. 
lst=list(map(int,input().split()))
my_set=set(lst)
lst1=list(my_set)
print(lst1)
```

```python
#Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị
#trí i), giữ nguyên thứ tự.
n=list(map(int,input().split()))
n=[0]+n
D=[0]*len(n)
for i in range(1, len(n)):
    D[i]=D[i-1]+n[i]
D.remove(0)
print(D)
```

```python
#Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple
#sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp.)
tpl=tuple(map(int,input().split()))
k=int(input())
k=k%len(tpl)
new_tpl=tpl[k:]+tpl[:k]
print(new_tpl)
```

```python
#Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách.
#Tạo dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện.
#Nếu key lặp lại, thêm vào list của key đó.
lst=list(input().split())
Dic={}
for i in lst:
    x, y= i.split(':')
    if x in Dic:
        Dic[x].append(y)
    else:
        Dic[x]=[y]
print(Dic)
```

```python
#Nhập một list số nguyên, In ra dictionary
#{'positives': ..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0)
lst=list(map(int,input().split()))
duong, am, ko=0, 0, 0
for i in lst:
    if i>0:
        duong+=1
    elif i<0:
        am+=1
    else:
        ko+=1
Dic={'positives':duong, 'negatives':am, 'zeros':ko}
print(Dic)
```

```python
#Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện của từng
#chuỗi và lưu vào dictionary. 
lst=list(input().split())
Dic={}
for i in lst:
    if i in Dic:
        Dic[i]+=1
    else:
        Dic[i]=1
print(Dic)
```

```python
#Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới
#trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có
#trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của key
def lst_dic(lst):
    Dic={}
    for i in lst:
        x, y=i.split(':')
        Dic[x]=int(y)
    return Dic
lst1=list(input().split())
lst2=list(input().split())
Dic1=lst_dic(lst1)
Dic2=lst_dic(lst2)
for i in Dic2:
    if i in Dic1:
        Dic1[i]+=Dic2[i]
    else:
        Dic1[i]=Dic2[i]
print(Dic1)
```

```python
#Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra
#list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple
#có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của
#số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên
#trong tuple.
lst=list(map(int,input().split()))
k=int(input())
Dic={}
my_set1=set(lst)
i=0
while i<len(lst):
    x=lst[i]
    y=k-x
    if y in my_set1:
        idx=lst.index(x)
        lst.remove(x)
        my_set2=set(lst)
        if y in my_set2 and x not in Dic:
            Dic[x]=y
        else:
            lst.insert(idx, x)
            i+=1
    else:
        i+=1
rst=[]
for i in Dic:
    rst.append((i,Dic[i]))
rst.sort()
print(rst)
```

```python
#Làm lại sau khi tham khảo chat gpt để độ phức tạp chỉ còn O(n).
#Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra
#list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple
#có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của
#số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên
#trong tuple.
lst=list(map(int,input().split()))
k=int(input())
seen=set()
pairs=set()
for x in lst:
    y=k-x
    if y in seen:
        pairs.add((y,x))
    seen.add(x)
my_lst=list(pairs)
my_lst.sort()
print(my_lst)
```

```python
#Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list.
#Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất
#trong số đó
lst=list(map(int,input().split()))
dic={}
for i in lst:
    if i not in dic:
        dic[i]=1
    else:
        dic[i]+=1
key=0
cnt=0
for i in dic:
    if (dic[i]>cnt) or (i<key and dic[i]==cnt):
        cnt=dic[i]
        key=i
print(key)
```

```python
#Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo một
#dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In
#ra dictionary mới. 
lst=list(input().split())
dic={}
for i in range(0, len(lst),2):
    dic[lst[i+1]]=lst[i]
print(dic)
```

```python
#Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các
#phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các
#phần tử chung đó, không trùng lặp
lst1=list(map(int,input().split()))
lst2=list(map(int,input().split()))
set1=set(lst1)
my_set=set()
for i in lst2:
    if i in set1:
        my_set.add(i)
my_lst=list(my_set)
print(my_lst)
```

```python
#Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số nguyên k ở dòng tiếp theo. 
#Hãy tạo một dictionary mới chỉ chứa các cặp keyvalue từ dictionary ban đầu mà có value lớn hơn k.
#In ra dictionary mới.
lst=list(input().split())
k=int(input())
dic={}
for i in range(1, len(lst), 2):
    x=int(lst[i])
    if x>k:
        dic[lst[i-1]]=x
print(dic)
```
