```python
#Viết hàm merge hai danh sách đã sắp xếp mà không dùng bất cứ thuật toán sắp xếp nào 
a=list(map(int,input().split()))
b=list(map(int,input().split()))
res=[]
i, j=0 ,0
while i<len(a) and j<len(b):
    if a[i]<b[j]:
        res.append(a[i])
        i+=1
    else:
        res.append(b[j])
        j+=1
res.extend(a[i:])
res.extend(b[j:])
print(*res)             
```

```python
## Cho ma trận hai chiều nxm có các phần tử được sắp xếp trên dòng tăng dần và trên 
#cột tăng dần, hãy tìm số lớn thứ k (k < nxm) với thuật toán tốt nhất có thể.
import heapq
m, n=map(int,input().split())
mtr=[]
for i in range(m):
    row=list(map(int,input().split()))
    mtr.append(row)
k=int(input())
lst=[]
heapq.heappush(lst, (mtr[0][0], 0, 0))
visited={(0, 0)}
x=0
cnt=0
i, j=0, 0
while cnt<k:
    x, i, j=heapq.heappop(lst)
    cnt+=1
    if i+1<m and (i+1, j) not in visited:
        heapq.heappush(lst, (mtr[i+1][j], i+1, j))
        visited.add((i+1, j))
    if j+1<n and (i, (j+1)) not in visited:
        heapq.heappush(lst, (mtr[i][j+1], i, j+1))
        visited.add((i, j+1))
print(x)
```

```python
## Cho ma trận hai chiều nxm có các phần tử được sắp xếp trên dòng tăng dần và trên 
#cột tăng dần, hãy tìm số lớn thứ k (k < nxm) với thuật toán tốt nhất có thể.
import heapq
m, n=map(int,input().split())
mtr=[]
for i in range(m):
    row=list(map(int,input().split()))
    mtr.append(row)
k=int(input())
lst=[]
for i in range(m-1, -1,-1):
    for j in range(n-1,-1,-1):
        if len(lst)<k:
           heapq.heappush(lst, -mtr[i][j])
        else:
            if mtr[i][j]<(-lst[0]):
                heapq.heappop(lst)
                heapq.heappush(lst, -mtr[i][j])
print(-lst[0])
```

```python
## Cho ma trận hai chiều nxm có các phần tử được sắp xếp trên dòng tăng dần và trên 
#cột tăng dần, hãy tìm số lớn thứ k (k < nxm) với thuật toán tốt nhất có thể.
def countLessEqual(mtr, m, n, mid):
    cnt=0
    i=m-1
    j=0
    while i>=0 and j<n:
        if mtr[i][j]<=mid:
            cnt+=i+1
            j+=1
        else:
            i-=1
    return cnt
def so_lon_thu_k(mtr, m, n, k):
    low=mtr[0][0]
    high=mtr[m-1][n-1]
    while low<high:
        mid=(low+high)//2
        if countLessEqual(mtr, m, n, mid) < k:
            low=mid+1
        else:
            high=mid
    return low
m, n=map(int,input().split())
mtr=[]
for i in range(m):
    row=list(map(int,input().split()))
    mtr.append(row)
k=int(input())
print(so_lon_thu_k(mtr, m, n, k))
            
```

```python
#Cho một số nguyên dương, hãy thực hiện hoán đổi đúng 2 chữ
#số để được một số nguyên nhỏ nhất có thể 
def giai(a):
    A=list(a)
    min_idx=len(A)-1

    swap_i=-1
    swap_j=-1
    for i in range(len(A)-2, -1, -1):
        if A[i]<A[min_idx]:
            min_idx=i
        else:
            swap_i=i
            swap_j=min_idx
    if swap_i==-1:
        A[-1], A[-2]=A[-2],A[-1]
    else:
        A[swap_i], A[swap_j]= A[swap_j], A[swap_i]
    
    kq=''
    for i in A:
        kq+=i
    return int(kq)
a=input()
print(giai(a))
```

```python
#Cho 2 xâu a,b, xác đính xem có cách biến đổi từ xâu a sang xâu b không. 
#Ghi ra "YES" nếu có và "NO" nếu ngược lại.
def check(a, b):
    cnt=0
    a2=''
    for i in a:
        if i.upper()==b[cnt]:
            a2+=i.upper()
            cnt+=1
    if a2==b:
        return True
    return False
n=int(input())
for i in range(n):
    a=input()
    b=input()
    if check(a,b):
        print('YES')
    else:
        print('NO')
```
