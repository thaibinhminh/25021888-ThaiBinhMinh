```python
#Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣
#trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải
def ma_tran(m,n):
    C=[]
    for i in range(m):
        row=[]
        for j in range(n):
            x=int(input())
            x=str(x)
            if len(x)<4:
                x=' '*(4-len(x))+x
            row.append(x)
        C.append(row)
    return C
m,n=map(int,input().split())
for i in ma_tran(m,n):
    print(i, end='\n')
```

```python
#Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và
#đường chéo phụ của mảng
def ma_tran_vuong(n):
    matrix=[]
    for i in range(n):
        row=list(map(int,input().split()))
        matrix.append(row)
    return matrix
n=int(input())
matrix_n=ma_tran_vuong(n)
cheo_chinh=[]
cheo_phu=[]
for i in range(n):
    cheo_chinh.append(matrix_n[i][i])
    cheo_phu.append(matrix_n[i][n-i-1])
print('Đường chéo chính', cheo_chinh)
print('Đường chéo phụ', cheo_phu)
```

```python
#Nhập vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên ̣ cùng một dòng,
#cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột k (hàng k) 
def ma_tran(m,n):
    matrix=[]
    for i in range(m):
        row=list(map(int,input().split()))
        matrix.append(row)
    return matrix
m, n, k=map(int,input().split())
ma_tran=ma_tran(m,n)
S=0
for i in range(m):
    S+=ma_tran[i][k-1]
print('Tổng các phần tử trên cột', k, 'là:', S)
```

```python
#hàm tìm số lớn nhất trong list
def max_of_list(lst):
    mx=-10**6
    for i in lst:
        if i>mx:
            mx=i
    return mx
lst=list(map(int,input().split()))
print(max_of_list(lst))
```

```python
def vi_tri(lst,k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1
lst=list(map(int,input().split()))
k=int(input())
print(vi_tri(lst,k))
```

```python
#vị trí có giá trị lớn nhất trong list
def vi_tri_co_gtln(lst):
    mx=-10**6
    x=0
    for i in range(len(lst)):
        if lst[i]>mx:
            mx=lst[i]
            x=i
    return x
lst=list(map(int,input().split()))
print(vi_tri_co_gtln(lst))
```

```python
def lst_khong_be_hon_x(lst,x):
    i=0
    while i<len(lst):
        if lst[i]<x:
            lst.pop(i)
        else:
            i+=1
    return lst
lst=list(map(int,input().split()))
x=int(input())
print(lst_khong_be_hon_x(lst,x))
```

```python
def day_0(lst):
    for i in range(len(lst)):
        if lst[i]==0:
            lst.pop(i)
            lst.insert(0,0)
    return lst
lst=list(map(int,input().split()))
print(day_0(lst))
```

```python
#Cho dãy chỉ gồm N giá trị từ 0 - 9 . Kiểm tra xem giá trị nào không có trong mảng
def so_thieu(n):
    my_set=set()
    for i in n:
        if i not in my_set:
            my_set.add(int(i))
    D=[]
    for i in range(10):
        if i not in my_set:
            D.append(i)
    return D
n=input()
print(so_thieu(n))
```

```python
#Tính tổng các phần tử cực đại địa phương trong một danh sách,
#không tính 2 phần tử đầu và cuối
def S_cddp(lst):
    S=0
    if len(lst)>3:
        for i in range(1,len(lst)-1):
            if lst[i-1]<lst[i]>lst[i+1]:
                S+=lst[i]
        return S
    elif len(lst)>2:
        if lst[0]<lst[i]>lst[2]:
            S=lst[i]
        return S
lst=list(map(int,input().split()))
print(S_cddp(lst))
```

```python
#Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước).
#Nếu không có in ra -1.
def be_hon_x(lst,x):
    D=[]
    for i in range (len(lst)):
        if lst[i]<x:
            D.append(i)
    if len(D)>0:
        return D
    else:
        return -1
lst=list(map(int,input().split()))
x=int(input())
print(be_hon_x(lst,x))
```

```python
#Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
#thực sự, hay vô danh (không tăng, không giảm)
def Tc_day(lst):
    giam=False
    tang=False
    for i in range (len(lst)-1):
        if lst[i]>lst[i+1]:
            giam=True
        elif lst[i]<lst[i+1]:
            tang=True
    if giam==tang:
        return 'vo danh'
    elif giam:
        return 'giam'
    else:
        return 'tang'
lst=list(map(int,input().split()))
print(Tc_day(lst))
```
