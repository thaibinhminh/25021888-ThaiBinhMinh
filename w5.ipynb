```python
#Kiểm tra số nguyên tố
import math
def so_nguyen_to(a):
    cnt=0
    for i in range(1, int(math.sqrt(a)+1)):
        if a%i==0:
            cnt+=1
    if a!=1 and cnt==1:
        print(a, 'là một số nguyên tố')
    else:
        print(a, 'không là một số nguyên tố')
a=int(input())
so_nguyen_to(a)
    
```

```python
#số hoàn hảo
def perfect_number(a):
    s=0
    for i in range(1,int(a**0.5)+1):
        if a%i==0:
            s+=i
        x=a//i
        if x!=a:
            s+=x
    if s==a:
        print('là số hoàn hảo')
    else:
        print('không là số hoàn hảo')
n=int(input())
perfect_number(n)
```

```python
#Viết hàm giả lập một máy tính bỏ túi.
#Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử\( operat \)
#Biết các toán tử đơn giản thực hiện phép toán lần lượt là:
#Phép trừ \( - \)
#Phép cộng \( + \)
#Phép chia \( / \)
#Phép nhân \( * \)
#Kết quả làm tròn đến chữ số thập phân thứ hai.
#Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)).
#Đầu ra: Kết quả phép toán. 
def phep_toan(a,b,n):
    if n=='+':
        return a+b
    elif n=='-':
        return a-b
    elif n=='*':
        return a*b
    else:
        return a/b
a, b=map(float,input().split())
n=input('Phép toán cần thực hiện(=,-,*,/): ')
print(f'{phep_toan(a,b,n):.2f}')
```

```python
#Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. 
#method 1:
def hamming1(a,b):
     return bin(a^b).count('1') #bin(): Chuyển số bất kì về chuỗi nhị phân
x1, y1=map(int,input().split())
print(hamming1(x1,y1))
#method 2:
def hamming2(a,b):
    x=a^b
    count=0
    while x:
        x=x&(x-1) #loại bỏ bit 1 thấp nhất, vd 1100 & 1011(x-1) = 1000
        count+=1
    return count
x2, y2=map(int,input().split())
print(hamming2(x2,y2))
```

```python
#Viết hàm trả về tổng các chữ số của số nguyên dương n 
def tong_chu_so(n):
    s=0
    while n>0:
        x=n%10
        s+=x
        n=n//10
    return s
n=int(input())
print(tong_chu_so(n))
```

```python
#Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). 
#Bạn có thể giả sử hai từ đầu vào có cùng độ dài
def dang_cau_1_phia(a,b):
    check=True
    b1=''
    for i in b:
        if i not in b1:
            b1+=i
    for i in b1:
        if b.count(i)!=1:
            s=0
            D=[]
            for j in range(b.count(i)):
                D.append(b.index(i,s))
                s=b.index(i,s)+1
            for k in range(len(D)-1):
                if a[D[k]]!=a[D[k+1]]:
                    check=False
    if check:
        return True
    else:
        return False
def dang_cau(a,b):
    if dang_cau_1_phia(a,b) and dang_cau_1_phia(b,a):
        return True
    else:
        return False
a=input()
b=input()
dang_cau(a,b)
```

```python
#Đẳng cấu( chat gpt ):
def dang_cau(a, b):
    # nếu độ dài khác thì chắc chắn không đẳng cấu
    if len(a) != len(b):
        return False
    #Tạo hai từ điển để lưu giá trị ánh xạ từ a sang b và ngược lại
    map_ab = {}
    map_ba = {}
    
    for x, y in zip(a, b):#Duyệt từng cặp phần tử x, y lần lượt trong a,b 
        if x in map_ab:
            if map_ab[x] != y:#Nếu x đã có trong từ điển mà có giá trị khác với y đang xét thì false
                return False
        else:
            map_ab[x] = y#Nếu x chx cs trong từ điển thì lưu x và giá trị khi ánh xạ từ a sang b

        if y in map_ba:
            if map_ba[y] != x:#Nếu y đã có trong từ điển mà có giá trị khác với x đang xét thì false
                return False
        else:
            map_ba[y] = x#Nếu y chx cs trong từ điển thì lưu y và giá trị khi ánh xạ từ b sang a
    return True
a=input()
b=input()
if dang_cau(a,b):
    print('là đẳng cấu')
else:
    print('Không là đẳng cấu')
```
