```python
#Kiểm tra và tính diện tích, chu vi của tam giác. 
a, b, c=map(float,input('Nhập độ dài 3 cạnh tam giác cách nhau bằng dấu cách: ').split())
if a+b>c or a+c>b or b+c>a:
    if a==b==c:
        print('Đây là 3 cạnh của một tam giác cân')
    elif a==b or b==c or a==c:
        print('Đây là 3 cạnh của tam giác cân')
    elif a*a==b*b+c*c or b*b==a*a+c*c or c*c==a*a+b*b:
        print('Đây là 3 cạnh của tam giác vuông')
    else:
        print('Đây là 3 cạnh của tam giác thường')
    p=(a+b+c)/2
    s=(p*(p-a)*(p-b)*(p-c))**0.5
    print(f'Diện tích tam giác là: {s: .2f}')
    print(f'Chu vi của tam giác là: {a+b+c: .2f}')
else:
    print('Đây không phải là ba cạnh của tam giác')
```

```python
#Giải phương trình bậc 1:
print('Để giải phương trình bậc 1 ax+b=0, bạn hãy nhập vào 2 số a và b')
a=float(input('Nhập a: '))
b=float(input('Nhập b: '))
if a!=0:
    x=-b/a
    print(f'Nghiệm của phương trình là: {x: .2f}')
else:
    print('Lỗi! Vui lòng nhập lại!')
```

```python
#Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?
x, y=map(float,input('Nhập tọa độ x,y của tâm đường tròn cách nhau bằng giấu cách: ').split(','))
r=float(input('Nhập bán kính của đường tròn: '))
x0, y0=map(float,input('Nhập tọa độ x0,y0 của điểm muốn kiểm tra có thuộc đường tròn hay không cách nhau bằng giấu cách: ').split(','))
import math
d=math.sqrt((x-x0)**2+(y-y0)**2)
if d==r:
    print('Điểm đó thuộc đường tròn.')
else:
    print('Điểm đó không thuộc đường tròn.')
```
