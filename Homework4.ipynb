```python
#VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi. Kiểm tra số này có phải là số nguyên tố hay không?

n=-1
while (n!=int(n)) or (n<=0) :
    n=float(input())
n=int(n)
if n==1:
    print('1 không phải là số nguyên tố')
else:
    check=True
    for i in range(2, int(n**(1/2))+1):
        if n%i==0:
            check=False
            break
if check:
    print(n, 'là số nguyên tố')
else:
    print(n, 'không là số nguyên tố') 
```

```python
#Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn [a, b]. Mỗi dòng là một bộ test (a b).
a=int(input())
b=int(input())
s=0
for i in range(a, b+1):
    if i!=1:
        check=True 
        for j in range(2, int(i**0.5)+1):
            if i%j==0:
                check=False
                break
        if check:
            s+=i
print(s)           
```

```python
#Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n.
n=int(input())
D=[]
for i in range(2,n+1):
    check=True
    for j in range(2, int(n**0.5)+1):
        if i%j==0:
            check=False
            break
    if check and n%i==0:
        D.append(i)
print(max(D))
```

```python
#Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.
n=int(input())
v=0
cnt=0
while str(n)!=str(n)[::-1]:
    cnt+=1
    while n>0:
        v=v*10+n%10
        n=n//10
    n=n+v
print(cnt)
print(n)
```

```python
#Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số 
#đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu 
#cách). Nếu không có, in ra (no number).
n=int(input())
check=True
for i in range(1,n+1):
    if (i**0.5)==int(i**0.5):
        print(i, end=' ')
        check=False
if check:
    print('no number')
```

```python
#Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất 
#và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số 
#nhỏ nhất.
n=int(input())
D=[]
for i in range (1,n+1):
    cnt=0
    while i!=1:
        if i%2==0:
            i=i//2
            cnt+=1
        else:
            i=i*3+1
            cnt+=1
    D.append(cnt)
print(max(D))
print(D.index(max(D))+1)
```

```python
#Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6.
n=int(input())
i=1
cnt=0
while i<int(n**0.5):
    if n%i==0:
        if i%2==0:
            cnt+=1
            x=n//i
        if x!=i and x%2==0:
            cnt+=1
    i+=1
print(cnt)
```

```python
#số thân thiết:
a, b=map(int,input().split())
def uoc_tru_chinh_no(x):
    s=0
    for i in range(1,int(x**0.5)+1):
        if x%i==0:
            s+=i
            y=x//i
            if y!=x:
                s+=y
    return s
if uoc_tru_chinh_no(a)==b and uoc_tru_chinh_no(b)==a:
    print('true')
else:
    print('false')       
```
