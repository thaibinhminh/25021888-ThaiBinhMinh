```python
a=input('Nhập một ký tự: ')
if len(a)!=1:
    print('Vui lòng nhập lại chỉ 1 ký tự')
else:
    if a.isdigit():
        print('Ký tự bạn nhập là số')
    elif a.isalpha():
        print('Ký tự bạn nhập là chữ cái')
    else:
        print('Ký tự bạn nhập không phải số cũng không phải chữ cái')
        
```
