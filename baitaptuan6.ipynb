```python
#Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng
#giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả. 
D=list(map(int,input().split()))
Y=[]
for i in D:
    if i not in Y:
        Y.append(i)
print(Y)
```

```python
#Nhập một dòng số nguyên. In ra list tổng tích lũy 
#(phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ tự.
D=list(map(int,input().split()))
D=[0]+D
S=[0]*len(D)
for i in range(1, len(D)):
    S[i]= S[i-1] + D[i]
X=S.pop(0)
print(S)
```

```python
#Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple sau khi xoay trái k vị trí. 
#(Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp.)
A=tuple(map(int,input().split()))
k=int(input())
X=[]
Y=[]
for i in A:
    if len(X)<k:
        X.append(i)
    else:
        Y.append(i)
B=Y+X
print(B)
```

```python
#Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
#dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key
#lặp lại, thêm vào list của key đó
X=input().split()
D={}
for i in X:
    key, value= i.split(':')
    if key not in D:
        D[key]=[value]
    else:
        D[key].append(value)
print(D)
```

```python
#Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0)
D=list(map(int,input().split()))
duong=0
am=0
cnt_0=0
for i in D:
    if i>0:
        duong+=1
    elif i<0:
        am+=1
    else:
        cnt_0+=1
Y={'positives':[duong],'negatives':[am],'zeros':[cnt_0]}
print(Y)
```

```python
#Viết chương trình nhập vào một list các chuỗi,
#sau đó đếm số lần xuất hiện của từng chuỗi và lưu vào dictionary.
lst=list(input().split())
D={}
for i in lst:
    if i not in D:
        D[i]=[lst.count(i)]
print(D)
```

```python
#Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một
#dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ
#được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới
#theo thứ tự từ điển tăng dần của key. 
a=list(input().split())
b=list(input().split())
c=a+b
D={}
for i in c:
    x, y= i.split(':')
    if x not in D:
        D[x]=y
    else:
        D[x]=int(D[x])+int(y)
print(D)
```

```python
#Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên
#dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là
#một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh
#sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được
#sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.
def so_can_tim(lst,x,y):
    lst1=list(lst)
    lst1.sort()
    lst1.remove(y)
    l=0
    r=len(lst1)-1
    while l<=r:
        mid=(l+r)//2
        if lst1[mid]==x:
            return True
            break
        elif lst1[mid]>x:
            r=mid-1
        else:
            l=mid+1
    return False
lst=list(map(int,input().split()))
k=int(input())
D=[]
i=0
while i<len(lst):
    y=lst[i]
    x=k-y
    if so_can_tim(lst,x,y):
        D.append((y,x))
        lst.remove(x)
        lst.remove(y)
    else:
        i+=1
D.sort()
print(D)
```

```python
#Tìm phần tử xuất hiện nhiều 
lst=list(map(int,input().split()))
D=[0]*len(lst)
lst.sort()
for i in range(len(lst)):
    D[i]=lst.count(lst[i])
x=lst[D.index(max(D))]
print(x)
```
